*Завдання 10. Аналіз індустрій серед високооплачуваних віддалених працівників

```python
#Інструмент, який потрібен для виконання цього завдання: Pandas (перцентилі, множинна фільтрація, підрахунок частоти)
#Опис завдання: Знайди респондентів, які:
#Знаходяться у 75 перцентилі за компенсацією (тобто заробляють більше, ніж 75% респондентів)
#Працюють віддалено
#Для цієї групи визнач найрозповсюдженіші індустрії.
```

```python
*Завдання 10. Аналіз індустрій серед високооплачуваних віддалених працівників
```

```python
*Завдання 10. Аналіз індустрій серед високооплачуваних віддалених працівників
import numpy as np
# видаляю строки, де відсутній запис в строкі компенсації
df_sorted_conv = df.dropna(subset = ['ConvertedCompYearly'])
# сортую строки, за зростанням в стовпчику компенсації
df_sorted_conv.sort_values(by='ConvertedCompYearly',inplace =True, ascending = False )
# знаходжу 75% перцентиль
percentile_75 = np.percentile(df_sorted_conv['ConvertedCompYearly'], 75)
```

```python
print(percentile_75)
```

```python
# вибираю тих респондентів, компенсація яких вище 75% прецентеля та які працюють віддалено
df_sorted_conv = df_sorted_conv.loc[(df_sorted_conv['ConvertedCompYearly'] > percentile_75) & (df_sorted_conv['RemoteWork'] == 'Remote')]
```

```python
df_sorted_conv['Industry'].mode()
```

```python
# групую респондентів по галузі, і підраховую кількість по кожній залузі
df_top = df_sorted_conv.groupby('Industry')['Industry'].count()
```

```python
#сортую значення частоти галузі по спаданню
df_top = df_top.sort_values(ascending= False)
```
